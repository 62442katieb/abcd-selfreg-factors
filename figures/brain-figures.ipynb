```python
from nilearn import plotting, surface
from os.path import join
import seaborn as sns
import nibabel as nib
```

```python
pc1 = 'pc1_tvals.nii.gz'
ttemp = 'ltransversetemporal_scaled.nii.gz'
```

```python
cmap = sns.choose_colorbrewer_palette('d', as_cmap=True)
```

```python
g = plotting.plot_glass_brain(pc1, vmin=3, vmax=3.5, cmap=cmap, colorbar=False, display_mode='lzry')
g.savefig('pc1-glass_brains.png', dpi=300)
```

```python
d = plotting.plot_glass_brain(ttemp, vmin=3, vmax=5, cmap=cmap, colorbar=True, display_mode='lzry')
d.savefig('sleep-glass_brains.png', dpi=300)
```

```python
q = plotting.plot_stat_map(pc1, cmap=cmap, threshold=3, cut_coords=[-4,16,-6], draw_cross=False)
```
